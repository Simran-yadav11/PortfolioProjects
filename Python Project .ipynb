# Zomato Dataset Analysis

# 1. Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

# 2. Reading CSV

```python
df=pd.read_csv(r"C:\Users\simran\OneDrive\Desktop\Data Science\Python\zomato.csv")
df.head()
```

```python
#Checking no. of rows and columns
df.shape
```

```python
#Checking columns
df.columns
```

```python
#Dropping columns that are not necessary
df = df.drop(['url','address','phone','dish_liked','menu_item','reviews_list'],axis=1)
df.head()
```

```python
#Checking null values in table
df.info()
```

# 3. Dropping Duplicates

```python
df.drop_duplicates(inplace = True)
df.shape
```

# 4. Cleaning Rate column

```python
df['rate'].unique()
```

# 5. Removing "New", "-" and "/5" from Rate column

```python
def handlerate(value):
    if(value=="NEW" or value=="-"):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
#Checking total null values in rate column
df.rate.isnull().sum()
```

# 6. Filling Null Values of Rate column with Mean

```python
df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

# 7. Dropping Null Values

```python
df.dropna(inplace = True)
df.head()
```

```python
df.info()
```

```python
df.rename(columns={'approx_cost(for two people)':'Cost2Plates','listed_in(type)':'Type'},inplace = True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

# 8. Both "listed_in(city)" and "location" are there, let's keep only one 

```python
df=df.drop(['listed_in(city)'],axis = 1)
```

```python
df
```

```python
df['Cost2Plates'].unique()
```

# 9. Removing , from Cost2Plate column 

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
df['Cost2Plates']=df['Cost2Plates'].apply(handlecomma)
df['Cost2Plates'].unique()
```

```python
df.head()
```

# 10. Cleaning "rest_type" column 

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

# 11. Making "rest_type_lessthan1000" in frequency as "others"

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

# 12. Cleaning "location" column

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending = True)
location_lessthan300 = location[location <300]
location_lessthan300
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
df['location']=df['location'].apply(handle_location)
df['location'].value_counts()
```

# 13. Cleaning "cuisines" column

```python
df['cuisines'].value_counts()
```

```python
cuisine = df['cuisines'].value_counts()
cuisine
```

```python
cuisine_lessthan100 = cuisine[cuisine<100]
cuisine_lessthan100
```

```python
def handle_cuisine(value):
    if(value in cuisine_lessthan100):
        return 'others'
    else:
        return value
df['cuisines'] = df['cuisines'].apply(handle_cuisine)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
df['Type'].value_counts()
```
